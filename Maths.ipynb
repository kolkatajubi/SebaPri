```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#uniform
#arr=np.random.random(1000)
#normal
arr=np.random.randn(1000)
#arr=np.asarray([9,1,6,2,90,77,13,6,9])
```

```python
class Statistics(object):
    def __init__(self,X):
        self.X=np.sort(X)        
    def std(self):
        return np.std(self.X)
    def mean(self):
        return np.mean(self.X)
    def median(self):
        n=len(self.X)
        if(n%2 == 0):
            return (self.X[int(n/2)]+self.X[int(n/2+1)])/2
        else:
            return self.X[int(n/2)]
    def freq_dist(self):
        return np.unique(self.X,return_counts=True)
    def mode(self):
        values,counts=self.freq_dist()
        return values[np.argwhere(counts==np.max(counts)).flatten()]
    def skew_type(self):
        dif=self.mean()-self.median()
        if dif>0.0001:
            return "Right Skewed"
        elif dif<-0.0001:
            return "Left Skewed"
        else:
            return "Normal"
    def iqr(self):
        n=len(self.X)
        q2=self.div(n)
        x=self.div(q2+1)
        q1=q2-x
        q3=q2+x
        return self.X[q3]-self.X[q1]
    def div(self,n):
        if(n%2 == 0):
            return int((int(n/2)+int(n/2+1))/2)
        else:
            return int(n/2)
            
```

```python
s= Statistics(arr)
fc=s.freq_dist()
std_pos=np.full(len(s.X),s.std()+s.mean())
mean=np.full(len(s.X),s.mean())
std_neg=np.full(len(s.X),s.mean()-s.std())
plt.show()
print(s.skew_type())
print("Mean :",s.mean())
print("Median :",s.median())
#print("Mode :",s.mode())
print("IQR :",s.iqr())
```

```python
plt.plot(s.X)
plt.plot(std_pos)
plt.plot(mean)
plt.plot(std_neg)
```
